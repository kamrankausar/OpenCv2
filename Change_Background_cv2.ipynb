```python
import matplotlib.pyplot as plt
import numpy as np
import cv2

%matplotlib inline
```

```python
# Read in and display the image
```

```python
#Read in the image
image = cv2.imread('pizza_bluescreen.jpg')
```

```python
#Print out the type of image data and its dimension(height, width and color)
print('This image is:', type(image),
      'with dimensions: ', image.shape)
```

```python
# Make a copy of the image
image_copy = np.copy(image)
```

```python
plt.imshow(image)
```

```python
#Change color to RGB (from BGR)
image_copy = cv2.cvtColor(image_copy, cv2.COLOR_RGB2BGR)
```

```python
#Display the image_copy
plt.imshow(image_copy)
```

```python
#Define the color threshold
```

```python
#TODO: Define the color selection boundaries in RGB values
# play around with these values until you isolate the blue background
lower_blue = np.array([0,0,200])
upper_blue = np.array([250,250,255])
```

```python
# Create a mask
```

```python
# Defone the masked area
mask = cv2.inRange(image_copy, lower_blue, upper_blue)
```

```python
# Vizualize the mask
plt.imshow(mask, cmap = 'gray')
```

```python
# Mask the image to let the pizza show through
masked_image = np.copy(image_copy)
masked_image[mask != 0] = [0,0,0]
```

```python
# Display it!
plt.imshow(masked_image)
```

```python
# Mask and add a background image
```

```python
# Load in a background image, and convert it to RGB
background_image = cv2.imread('space_background.jpg')
background_image = cv2.cvtColor(background_image, cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(background_image)
```

```python
# Crop it to the right size(720, 1280)
crop_background = background_image[0:720, 0:1280]
```

```python
# Mask the cropped background so that the pizza area is blocked
crop_background[mask == 0] = [0,0,0]
#Display the background
plt.imshow(crop_background)
```

```python
# Create a complete image
# Add the two images together to create a complte image
complete_image = masked_image + crop_background
```

```python
#Display the result
plt.imshow(complete_image)
```
